```python
def obtener_cuotas():
    from bs4 import BeautifulSoup
    import requests
    import pandas as pd

    # URL inicial
    url = 'https://apuestas.wplay.co/es/s/FOOT/F%C3%BAtbol'

    # Obtener el HTML
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'lxml')

    # Obtener el contenedor principal
    container = soup.find('div', id='main-area')

    # Encontrar todos los fragmentos de cada liga
    fragmentos_ligas = container.find_all('div', class_='fragment expander coupon-for-type')

    partidos_totales = []

    # Recorrer cada fragmento de liga y extraer los datos
    for fragmento_liga in fragmentos_ligas:
        # Encontrar el título de la liga
        titulo_liga = fragmento_liga.find_previous('h4', class_='expander-button').text.strip()

        # Buscar los elementos span con la clase 'seln-name' y crear una lista con su texto
        equipos = [elem.text for elem in fragmento_liga.find_all('span', class_='seln-name')]

        # Buscar los elementos span con la clase 'price dec' y crear una lista con su texto
        puntos = [elem.text for elem in fragmento_liga.find_all('span', class_='price dec')]

        # Dividir la lista de equipos en dos: locales y visitantes
        equipos_locales = equipos[::2]
        equipos_visitantes = equipos[1::2]

        # Asegurarse de que las listas tienen la misma longitud
        assert len(equipos_locales) == len(equipos_visitantes)

        # Dividir la lista de puntos en grupos de 3 (cuotas para cada partido)
        assert len(puntos) % 3 == 0
        cuotas = [puntos[i:i+3] for i in range(0, len(puntos), 3)]

        # Crear una lista de tuplas con la información de cada partido
        partidos_liga = []
        for local, visitante, cuotas_partido in zip(equipos_locales, equipos_visitantes, cuotas):
            cuota_local, cuota_empate, cuota_visitante = cuotas_partido
            partido = (local, visitante, cuota_local, cuota_empate, cuota_visitante, titulo_liga)
            partidos_liga.append(partido)

        partidos_totales.extend(partidos_liga)

    # Crear el DataFrame
    df = pd.DataFrame(partidos_totales, columns=['local', 'visitante', 'cuota_local', 'cuota_empate', 'cuota_visitante', 'liga'])

    # Guardar en un archivo Excel
    writer = pd.ExcelWriter('cuotas_partidos.xlsx', engine='xlsxwriter')
    df.to_excel(writer, sheet_name='Cuotas_Partidos', index=False)
    writer.save()

    return df
```
